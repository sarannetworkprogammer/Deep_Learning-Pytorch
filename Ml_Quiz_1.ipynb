```python
# Q6: Replace the value 2 from column "sex" with 0. Initially 1 stands for male and 2 for female. 

df2 = df1.replace({'sex':2}, 0)
df2


```

```python
# Q7: Find all the males with whose age is between 30 to 40

df2[((df2.age >=30) & (df2.age <= 40)) & (df2.sex == 1)]
```

```python
# Q8: Remove the columns "wtkg2" and "wtyrago"

df2.drop(['wtkg2','wtyrago'], axis=1)

```

```python
# Q9: Rename the column "weight2" with "weight" and "htm3" with "height".


df3 = df2.rename(columns={'weight2':'weight', 'htm3':'height'})
df3

```

```python
# Q10: Find all the females with height greater than 160 cm.


df3[(df3.height >= 160)  & (df3.sex == 0)]

```
